# Preprocessing

```python
import re
import pandas as pd
```

```python
f = open("sample.txt", 'r', encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[APap][Mm]\s-\s'
```

```python
# first string was an empty string
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python

# Create DataFrame
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

# Replace the narrow no-break space with a regular space using regex
df['message_date'] = df['message_date'].str.replace(r'\u202f', ' ', regex=True)

# Convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M %p - ', errors='coerce')

# Rename the column
df.rename(columns={'message_date': 'date'}, inplace=True)
df
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
df.head()
```

```python
df['year'] = df['date'].dt.year
```

```python
df.head()
```

```python
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
```

```python
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```
